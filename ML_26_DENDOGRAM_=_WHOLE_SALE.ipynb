```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/datasets/Wholesale customers data.csv')
df
```

```python
#to scale from zero to 1
from sklearn.preprocessing import normalize
data=normalize(df)
data
```

```python
new_data=pd.DataFrame(data,columns=df.columns)
new_data
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.title('DENDROGRAM')
dend=shc.dendrogram(shc.linkage(new_data,method='single'))
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.title('DENDROGRAM')
dend=shc.dendrogram(shc.linkage(new_data,method='average'))
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.title('DENDROGRAM')
dend=shc.dendrogram(shc.linkage(new_data,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
cluster.fit_predict(new_data)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(new_data['Milk'],new_data['Grocery'],c=cluster.labels_  )
```
